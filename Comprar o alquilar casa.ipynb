```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
import seaborn as sb
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_selection import SelectKBest
```

```python
df = pd.read_csv(r"comprar_alquilar.csv")
df.head(10)
```

```python
print(df.groupby('comprar').size())
```

```python
df.drop(['comprar'], axis=1).hist()
plt.show()
```

```python
df['gastos']=(df['gastos_comunes'] + df['gastos_otros'] + df['pago_coche'])
df['financiar'] = df['vivienda'] - df['ahorros']
df.drop(['gastos_comunes', 'gastos_otros', 'pago_coche'], axis = 1).head(10) #Lo que realizamos es agrupar diferentes columnas
#Tambin creamos columna financiar.
```

```python
#Revisamos un resumen estadstico gracias a pandas y describe():
reduced = df.drop (['gastos_comunes','gastos_otros', 'pago_coche'], axis = 1)
reduced.describe()
```

```python
#En vez de utilizar las 11 COLUMNAS de datos de entrada, con SKLEARN (SELECTKBEST) seleccionaremos
#las 5 mejores caractersticas y usaremos slo esas.

X = df.drop(['comprar'], axis = 1)
y = df['comprar']

best=SelectKBest(k=5)
X_new = best.fit_transform(X,y)
X_new.shape
selected = best.get_support(indices=True)
```

```python
#En esta parte utilizamos 5 de las 11 features(caracteristicas) que
#Seleccionamos arriba gracias a SelectKBest y son las que ms "aportan".
used_features = X.columns[selected]

colormap = plt.cm.viridis
plt.figure(figsize=(12,12))
plt.title('Caracteristicas Correlacin de pearson', y=1.05, size=15)
sb.heatmap(df[used_features].astype(float).corr(),linewidths=0.1,vmax=1.0,square=True,
           cmap=colormap,linecolor='white',annot=True)
```

```python
#Procedemos a crear el modelo Gaussian NB con SKLearn
#Dividiremos nuestros datos de entrada en entrenamiento y test

X_train,X_test = train_test_split(df,test_size=0.2,random_state=6)
y_train=X_train['comprar']
y_test=X_test['comprar']
```

```python
#Creamos el modelo y lo entrenamos con FIT() y
#Obtenemos predicciones sobre nuestro datatest

gnb=GaussianNB() #Inicializamos nuestro clasificador

#entrenamos el clasificador
gnb.fit(X_train[used_features].values, y_train)
result=gnb.predict(X_test[used_features])
print('precisin en el set de entrenamiento: {:.2f}'.format(gnb.score(X_train[used_features], y_train)))
print('Precisin en el set de test: {:.2f}'.format(gnb.score(X_test[used_features], y_test)))
```

```python
import pickle

#Definimos funcion de guardar y cargar
def save_object(filename, object):
    with open(''+filename, 'wb') as file:
        pickle.dump(object, file)
def load_object(filename):
    with open(''+filename, 'rb') as f:
        loaded = pickle.load(f)
    return loaded
#Guardar los objetos que necesitaremos mas adelante
save_object('comprar_alquilar.pkl',gnb)
#Cargamos cuando haga falta
load = load_object('comprar_alquilar.pkl')
```
